```python
import os
import json
from doctr.io import DocumentFile
from doctr.models import ocr_predictor

# Initialize the OCR model
model = ocr_predictor(det_arch='db_resnet50', reco_arch='vitstr_small', pretrained=True)

# Define paths
input_base = "/kaggle/input/find-it-again-dataset/findit2"  # Update this to your dataset path
output_base = "/kaggle/working/doctr-output"
subfolders = ["train", "test", "val"]

def process_folder(input_folder, output_folder):
    """Process all images in a folder and save results"""
    os.makedirs(output_folder, exist_ok=True)
    
    for img_file in os.listdir(input_folder):
        if not img_file.lower().endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp', '.gif')):
            continue
            
        img_path = os.path.join(input_folder, img_file)
        base_name = os.path.splitext(img_file)[0]
        
        try:
            # Process image
            doc = DocumentFile.from_images(img_path)
            result = model(doc)
            
            # Save JSON
            json_path = os.path.join(output_folder, f"{base_name}.json")
            with open(json_path, 'w') as f:
                json.dump(result.export(), f, indent=2)
            
            # Save text
            txt_path = os.path.join(output_folder, f"{base_name}.txt")
            with open(txt_path, 'w') as f:
                f.write(result.render())
            
            print(f"Processed: {img_path}")
            
        except Exception as e:
            print(f"Error processing {img_path}: {str(e)}")

# Process all folders
for folder in subfolders:
    input_dir = os.path.join(input_base, folder)
    output_dir = os.path.join(output_base, folder)
    print(f"Processing {input_dir}...")
    process_folder(input_dir, output_dir)

print("Processing completed!")
```
